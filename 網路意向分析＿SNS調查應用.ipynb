```python
def GetOpinionData(opinionFile):
    print('即將要打開', opinionFile)
    ret_lines = ""
    with open(opinionFile, 'r') as read_file:
        lines = read_file.readlines()
    for line in lines:
        ret_lines += line
        
    print('Note ==>檔案已經讀完')
    return ret_lines
```

```python
dict_SNS = {
    'Facebook':['facebook', 'fb', '臉書'],
    'Instagram':['instagram', 'ig'],
    'Line':['line'],
    'Twitter':['twitter'],
    'Youtube':['youtube'],
    '巴哈姆特':['巴哈姆特'],
    'Dcard':['dcard'],
    'Google':['google'],
    'PTT':['ptt'],
    'Moodle':['moodle'],
    'Medium':['medium'],
    'LetterBoxed':['letterBoxed'],    
}

#print(dict_SNS)

str_opinion = GetOpinionData('collected_opinion.txt')
str_opinion = str_opinion.lower()
#print(str_opinion)

dict_statSNS = {}

for k_sns in dict_SNS:
    #print(k_sns, '=>', dict_SNS[k_sns])
    count_k_sns = 0
    for callname in dict_SNS[k_sns]:
        #print('\t, callname')
        #print(str_opinion.count(callname))
        count_k_sns += str_opinion.count(callname)
    print(k_sns, '出現了', count_k_sns)
    dict_statSNS[k_sns] = count_k_sns

print(dict_statSNS)
```

```python
plt.figure(figsize=(18,9))

y = dict_statSNS.keys()
size = len(dict_statSNS)
x = dict_statSNS.values()

my_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'darkblue', 'purple']

plt.pie(x,
       labels = y,
       autopct = '%1.1f%%',
       pctdistance = 0.6,
       textprops = {"fontsize" : 12},
       shadow=True)

plt.axis('equal')
plt.title("本班同學參與SNS概況", {"fontsize" : 18})
plt.legend(loc = "best")

plt.show()
```
